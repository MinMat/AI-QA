## Load model

```python
# Load model
from models import InferSent
model_version = 1
MODEL_PATH = "../encoder/infersent%s.pkl" % model_version
params_model = {'bsize': 64, 'word_emb_dim': 300, 'enc_lstm_dim': 2048,
                'pool_type': 'max', 'dpout_model': 0.0, 'version': model_version}
model = InferSent(params_model)
model.load_state_dict(torch.load(MODEL_PATH))
```

```python
# Keep it on CPU or put it on GPU
use_cuda = False
model = model.cuda() if use_cuda else model
```

```python
# If infersent1 -> use GloVe embeddings. If infersent2 -> use InferSent embeddings.
W2V_PATH = '../dataset/GloVe/glove.840B.300d.txt' if model_version == 1 else '../dataset/fastText/crawl-300d-2M.vec'
model.set_w2v_path(W2V_PATH)
```

```python
question_stem = list(train["Question Stem"])    
```

```python
len(question_stem)
```

```python
blob = TextBlob(" ".join(question_stem))
question_sentences = [item.raw for item in blob.sentences]
len(question_sentences)
#print(question_sentences)
```

```python
dict_embeddings = {}
for i in range(len(sentences)):
    dict_embeddings[sentences[i]] = model.encode([sentences[i]], tokenize=True)
```

```python
dict_embeddings_questions = {}
for i in range(len(question_sentences)):
    dict_embeddings_questions[question_sentences[i]] = model.encode([question_sentences[i]], tokenize=True)
```

```python
def cosine(u, v):
    return np.dot(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
```

```python
cosine(model.encode(['Frilled sharks and angler fish live far beneath the surface of the ocean, which is why they are known as.'])[0], model.encode(['deep sea animals live deep in the ocean'])[0])
```

```python
cosine(model.encode(['beneath the surface of the ocean'])[0], model.encode(['deep in the ocean'])[0])
```
